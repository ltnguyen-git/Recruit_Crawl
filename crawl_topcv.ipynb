```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import requests
import pandas as pd
import math
print("Importing done")
```

```python
driver = webdriver.Chrome()
url = 'https://www.topcv.vn/'
driver.get(url)
sleep(3)
```

```python
key_word = 'Data Analyst'
search_job = driver.find_element('xpath','//*[@id="keyword"]' )
search_job.send_keys(key_word)

search_button = driver.find_element('xpath','//*[@id="frm-search-job"]/div/div[2]/button')
search_button.click()
sleep(1)

```

```python
sleep(2)
non_ads = driver.find_element('xpath', '//*[@id="modal-tet-event"]/div/div/div[1]/button')
non_ads.click()
```

```python
number_of_jobs = driver.find_element('xpath', '//*[@id="main"]/div[1]/div[3]/div[2]/div[1]/span/b').text
page = math.ceil(int(number_of_jobs)/25)
print(page) 
print(number_of_jobs)
arr_link_page = []
link= []
card = []
for i in range(1, page +1):
    link_page = f'https://www.topcv.vn/tim-viec-lam-{key_word.lower().replace(" ", "-")}?&page={i}'
    arr_link_page.append(link_page)

title =[]
company =[]
company_website =[]
for link_page in arr_link_page:
    driver.get(link_page)
    requests.get(link_page)
    soup =  BeautifulSoup(driver.page_source)
    tag_h3 = (soup.find_all('div','body'))
    sleep(1)
    print(len(tag_h3))
    print(link_page)
    for tag_a in tag_h3:
        title.append(tag_a.find('span').text)
        try:
            tag = tag_a.find('a').get('href')
            card.append(tag)
        except:
            card.append('nan')
        
        company.append(tag_a.find('a','company').text)
        company_website.append(tag_a.find('a','company').get('href'))


```

```python
salary =[]
number_need = []
working_form = []
genders = []
position = []
experience = []
locations = []
job_description = []
```

```python
get_data =[]
des =[]
for links in card:
   driver.get(links)
   page_source = driver.page_source
   soup =  BeautifulSoup(page_source, 'html.parser')


   tag_body = soup.find_all('div','box-main')

   for tag in tag_body:
      tg = tag.find_all('div', 'box-item')
      for t in tg:
         need =  t.find('span').text
         get_data.append(need)

   char_remove = ['\n- ',"- ", '\n']
   tag_location = soup.find_all('div','box-address')
   for t in tag_location:
      k =  t.find('div').text
   # loc = tag_location.get('div')
      for char in char_remove:
         c = k.replace(char, "")
      locations.append(c)

   tag_des = soup.find_all('div', 'job-data')
   for n in tag_des:
      d = n.find_all('div','content-tab')
      for p in d:
         q = p.text
         des.append(q)
```

```python
fes=[]
job_description = []
for i in range(0, len(des),3):
    job_description.append(str(des[i]) + str(des[i+1]) + str(des[i+2]))

print(job_description[:1])
print(len(job_description))
char_des_remove = ['.\xa0•', '•', '\xa0•','.\n','\n']
for t in job_description:
    # print(t)
    for char_des in char_des_remove:
        k = (t).replace(char_des, ' ')
    fes.append(k)

print(len(fes))
print(fes[:1])
```

```python
for i in range(0, len(get_data),6):
    salary.append(get_data[i])
    number_need.append(get_data[i+1])
    working_form.append(get_data[i+2])
    position.append(get_data[i+3])
    genders.append(get_data[i+4])
    experience.append(get_data[i+5])
```

```python
df =pd.DataFrame({'title':title,
                  'name_company':company,
                  'company_website':company_website,
                  'working_form' : working_form,
                  'position': position,
                  'gender':genders,
                  'experience':experience,
                  'location':locations,
                  'link_job':card,
                  'job_description': fes,
                 }, columns = ['title','name_company','company_website','working_form','position','gender','experience','location','link_job','job_description'])
```
