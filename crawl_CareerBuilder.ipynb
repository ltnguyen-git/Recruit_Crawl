```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import requests
import pandas as pd
from math import *
import math
print("Importing done")
```

```python
driver = webdriver.Chrome()
url = 'https://careerbuilder.vn/'
driver.get(url)
sleep(8)
```

```python
key_words = 'Data Analysis'
input_fields = driver.find_element('xpath', '//*[@id="keyword"]')
input_fields.send_keys(key_words)
sleep(1)
```

```python
find_job_buttons =driver.find_element('xpath', '/html/body/main/div[1]/section[2]/div/div/div[1]/form/div[4]/button')
find_job_buttons.click()
sleep(3)
```

```python
count_jobs  = driver.find_element('xpath', '/html/body/main/section[2]/div/div/div[1]/div[1]/div[1]/h1').text
print(count_jobs)
print(count_jobs.split())
for s in count_jobs.split():
    if s.isdigit():
        print(s)
        number_of_page = math.ceil(int(s)/50)


print(number_of_page)



# len(link)
```

```python
headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}

links=[]
for i in range(1, number_of_page+1):
    link_all_page= f'https://careerbuilder.vn/viec-lam/Data-Analysis-k-trang-{i}-vi.html'
    driver.get(link_all_page)
    sleep(1)
    r = requests.get(link_all_page, headers= headers)
    soup = BeautifulSoup(driver.page_source)
    tag = soup.find_all(name='div', attrs={'class':'job-item'})
    for t in tag:
        a = t.find(name = 'a', attrs = {'class' : 'job_link'})
        links.append(a['href'])
```

```python
len(links)
```

```python
links
```

```python
title = []
company = []
website_company = []
list_links = []
```

```python
lo = []
working_form =[]
groups = []
positions = []
salary = []
experiences = []
locations = []
for link_job in links:
    driver.get(link_job)
    sleep(1)
    soup = BeautifulSoup(driver.page_source)

    tag_div = soup.find_all(name='div',attrs='col-12 mb-15')
    if tag_div is not None:
        for t in tag_div:
            
            t1 = t.find(name='h1',attrs='title').text
            title.append(t1)
            list_links.append(link_job)
            try:
                tg = t.find(name='a',attrs='employer job-company-name')
                company.append(tg.text)
                website_company.append(tg['href'])
            except:
                company.append('nan')
                website_company.append('nan')
                
            s = driver.find_element('xpath','//*[@id="tab-1"]/section/div[1]/div/div[3]/div/ul/li[1]/p')
            salary.append(s.text)

            l = driver.find_element('xpath','/html/body/main/section[2]/div/div/div[2]/div/div[1]/section/div[1]/div/div[1]/div/div/p/a')
            locations.append(l.text)

            wf = driver.find_element('xpath','/html/body/main/section[2]/div/div/div[2]/div/div[1]/section/div[1]/div/div[2]/div/ul/li[3]/p')
            working_form.append(wf.text)

            p = driver.find_element('xpath','/html/body/main/section[2]/div/div/div[2]/div/div[1]/section/div[1]/div/div[3]/div/ul/li[3]/p')
            positions.append(p.text)

            e = driver.find_element('xpath','/html/body/main/section[2]/div/div/div[2]/div/div[1]/section/div[1]/div/div[3]/div/ul/li[2]/p')
            experiences.append(e.text)

            g = driver.find_element('xpath','/html/body/main/section[2]/div/div/div[2]/div/div[1]/section/div[1]/div/div[2]/div/ul/li[2]/p')
            groups.append(g.text)  


```

```python
df =pd.DataFrame({'title':title,
                  'name_company':company,
                  'company_website':website_company,
                  'working_form' : working_form,
                  'position': positions,
                  'groups':groups,
                  'experience':experiences,
                  'location':locations,
                  'link_job':list_links,
                #   'job_description': fes,
                 }, columns = ['title','name_company','company_website','working_form','position','experience','location','groups','link_job'])
```
