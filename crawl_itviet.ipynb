```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import requests
import pandas as pd
import math
print("Importing done")
```

```python
driver = webdriver.Chrome()
url = 'https://itviec.com/sign_in'
driver.get(url)
sleep(3)
```

```python
credential = open('account.txt')
line = credential.readlines()
username = line[0]
password = line[2]

print('Finish importing the login credentials')
```

```python
# login_button = driver.find_element('xpath', '//*[@id="navbarNavDropdown"]/ul[2]/li[2]/a')
# login_button.click()
# sleep(1)

email_field = driver.find_element('xpath', '//*[@id="user_email"]')
email_field.send_keys(username)
sleep(1)

password_field = driver.find_element('xpath', '//*[@id="user_password"]')
password_field.send_keys(password)
sleep(1)

login_in = driver.find_element('xpath', '//*[@id="container"]/div[3]/div/div[2]/form/div[4]/div/button')
login_in.click()
sleep(3)

```

```python
try: 
    non_ads = driver.find_element('xpath', '//*[@id="upload-cv-modal"]/div/div/div/button')
    non_ads.click()
except:
    pass

city = driver.find_element('xpath', '//*[@id="search_form"]/div[2]/div/div')
city.click()

all_city = driver.find_element('xpath', '//*[@id="city-ts-control"]')
all_city.click()

key_word = 'data'
search_job = driver.find_element('xpath', '//*[@id="query"]')
search_job.send_keys(key_word)

ran_click = driver.find_element('xpath', '//*[@id="search_form"]/div[4]')
ran_click.click()
sleep(1)
search_button = driver.find_element('xpath', '//*[@id="search_form"]/div[5]')
search_button.click()
```

```python
number_jobs = driver.find_element('xpath', '//*[@id="jobs"]/h1').text
print(number_jobs)

for s in number_jobs.split():
    if s.isdigit():
        print(s)
        number_of_page = math.ceil(int(s)/20)

print(number_of_page)

tag_h3 = []
cards = []
arr_all_pages = []

for i in range(1,number_of_page+1):
    link_all_pages = f'https://itviec.com/it-jobs/data?job_selected={key_word}&page={i}'
    arr_all_pages.append(link_all_pages)

print(arr_all_pages)

list_names = []
list_links = []

for link_page in arr_all_pages:
    driver.get(link_page)
    requests.get(link_page)
    soup =  BeautifulSoup(driver.page_source)
    tag_h3 = (soup.find_all('h3', 'title job-details-link-wrapper'))
    sleep(1)
    for tag_a in tag_h3:
        tags = tag_a.find(name ='a')
        list_names.append(tags.text.strip())
        list_links.append('https://itviec.com/'+ tags['href'])

    
```

```python
# list name có ở trên
company = []
website_company = []
address_company = []
work_at = []
salary= []
skills_require= []
# list link có ở trên
a=[]
```

```python
for job_link in list_links:
    driver.get(job_link)
    sleep(3)
    page_source = driver.page_source
    soup =  BeautifulSoup(page_source, 'html.parser')

    try:
        # company.append(driver.find_element('xpath', '//*[@id="container"]/div[3]/div[3]/div/div/div[1]/div[2]/div').text)
        company_names = soup.find_all(name = 'h3', attrs = {'class' : 'employer-long-overview__name hidden-xs d-none d-sm-block'})

        for name in company_names:
            job = name.find('a')
            company.append(job.text.strip())
    except:
        company.append('nan')
    
    try:
        company_website = soup.find_all('h3', 'employer-long-overview__name hidden-xs d-none d-sm-block' )
        for web in company_website:
            website = web.find('a')
            website_company.append('https://itviec.com/'+ website['href'])
        # website_company.append(driver.find_element('xpath', '//*[@id="container"]/div[3]/div[3]/div/div/div[1]/div[2]/h3/a').get_attribute("href"))
    except:
        website_company.append('nan')
    
    try:
        # add = soup.find_all('div','svg-icon')
        # for a in add:
        #     adt = a.find('div','svg-icon__text')
        #     address_company.append(adt.text().strip())
        address_company.append(driver.find_element('xpath', '//*[@id="container"]/div[3]/div[2]/div/div[2]/div[3]/div/span').text)

    except:
        address_company.append('nan')
    
    try:
        # work_where = soup.find_all('div', 'svg-icon__box d-flex align-items-center')
        # for w in work_where:
        #     ww = w.find('div','svg-icon__text')
        #     work_at.append(ww.text.strip())
        work_at.append(driver.find_element('xpath', '//*[@id="container"]/div[3]/div[2]/div/div[2]/div[4]/div/div/span').text)
    except:
        work_at.append('nan')
    

    # try:
    #     # skill_test = soup.find_all('div', 'job-details__tag-list')
    #     skill_test = soup.find_all('a','big ilabel mkt-track').get('span')
    #     # for list_skill in skill_test:
    #     #     a.append(list_skill.get('span'))
    #     skills_require.append(a)
    # except: 
    #     skills_require.append('nan')

    a = []
    page_response = requests.get(job_link)
    page_content = BeautifulSoup(markup=page_response.content, features="html.parser")

    skill_require = page_content.find_all(name = 'a', attrs = {'class' : 'big ilabel mkt-track'})

    for skill in skill_require:
        a.append(skill.text.replace('\n', ''))

    skills_require.append(a)


    try:
        # luong = soup.find_all('div', 'svg-icon svg-icon--rounded svg-icon--green')
        # for s in luong:
        #     sala= s.find('div','svg-icon__text')
        #     salary.append(sala.text.strip())
        salary.append(driver.find_element('xpath', '//*[@id="container"]/div[3]/div[2]/div/div[2]/div[2]/div').text)
    except:
        salary.append('nan')

```

```python
df =pd.DataFrame({
                  'name_job':list_names,
                  'company':company,
                  'company_website':website_company,
                  'address_company' : address_company,
                  'work_at': work_at,
                  'salary':salary,
                  'skills':skills_require,
                  'link_job': list_links,
                 }, columns = ['name_job', 'company', 'company_website','address_company', 'work_at', 'salary','skills', 'link_job'])
```
